```python
def sigmoid(y):
    return (1/(1+np.exp(-y)))
```

```python
def p(x,b0,b1):
    return (sigmoid(b0 + b1*x))
```

```python
def cost(x,y,b0,b1):
    n = len(x)
    c = 0
    for i in range(n):
        c = c+y[i]*np.log(p(x[i],b0,b1) + 0.00001) + (1-y[i]) * np.log(1 - p(x[i],b0,b1)+0.0001)
    return -1*(c/n)
```

```python
def gradientDecent(x, y, epoch):
    lamb = 0.0001
    b0 = 3.58
    b1 = 2.72

    counter = 0
    c = cost(x,y,b0,b1)

    while(True):
        c = cost(x,y,b0,b1)
        b0_cost = 0
        b1_cost = 0
        for i in range(len(x)):
            b1_cost = b1_cost + (-y[i] + p(x[i],b0,b1)) * x[i]
            b0_cost = b0_cost + (-y[i] + p(x[i],b0,b1))
        b1 = b1 - (1/len(x)) * lamb * b1_cost
        b0 = b0 - (1/len(x)) * lamb * b0_cost
        c1 = cost(x,y,b0,b1)

        counter += 1
        print(cost(x,y,b0,b1))

        if(abs(c1-c) < 0.0000001 or epoch == counter):
            return (b0, b1)
```

```python
b0,b1 = gradientDecent(x,y,1000)
```

```python
print(b0,b1)
```

```python
def yPredict(b0,b1,x):
    l = []
    for i in range(len(x)):
        if(p(x[i],b0,b1) >= 0.5):
            l.append(1)
        else:
            l.append(0)
    return(np.array(l))
```

```python
y_predict = yPredict(b0,b1,x)
```
