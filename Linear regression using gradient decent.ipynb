```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
def ypred(slope,intercept,x):
    yp=intercept+x*slope
    return yp
```

```python
def cost(x,y,sl,inter):
    n=len(x)
    sum=0
    for i in range(n):
        yp=ypred(sl,inter,x[i])
        sum=sum+(yp-y[i])**2 #calculates the squared error between the predicted value (yp) and the actual value
    c=0.5*sum/n
    return c
    
```

```python
def errIntercept(x,y,sl,inter):
    sum=0
    n=len(x)
    for i in range(n):
        yp=ypred(sl,inter,x[i])
        sum=sum+(yp-y[i])
    return sum
    
```

```python
def errSlope(x,y,sl,inter):
    sum=0
    n=len(x)
    for i in range(n):
        yp=ypred(sl,inter,x[i])
        sum=sum+(yp-y[i])*x[i]
    return sum
```

```python
def gd(x,y,lr):
    slope=1;
    intercept=100
    c=cost(x,y,slope,intercept)
    #while c>0.00005:
    for epoch in range(500):
        intercept=intercept-lr*errIntercept(x,y,slope,intercept)/len(x)
        slope=slope-lr*errSlope(x,y,slope,intercept)/len(x)
        print("Epoch ",epoch,slope,intercept)
        c=cost(x,y,slope,intercept) 
    return slope,intercept
```

```python
sl,inter=gd(x,y,0.01)
```

```python
print(sl,inter)
```

```python
yp=[]
for i in range(len(x)):
    yi=ypred(sl,inter,x[i])
    yp.append(yi)
```

```python
plt.scatter(x,y)
plt.plot(x,yp,c='r')
plt.show()
```

```python
def sse(x,y,sl,inter): #sum of squared error
    n=len(x)
    sum=0
    for i in range(n):
        yp=ypred(sl,inter,x[i])
        sum=sum+(yp-y[i])**2
    return sum
```

```python
def varErr(y): #Total varience
    ymean=np.mean(y)
    n=len(y)
    sum=0
    for i in range(n):
        sum=sum+(y[i]-ymean)**2
    return sum
```

```python
def rSquared(x,y,slope,intercept):
    r=1-sse(x,y,slope,intercept)/ varErr(y)
    return r   
```
